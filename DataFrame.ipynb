Persona altura mínima


```python
import pandas as pd
gent = pd.DataFrame([{"Nombre":"Pedro","Edad":25,"Altura":1.68},
                     {"Nombre":"Sonia","Edad":52,"Altura":1.74},
                     {"Nombre":"Lluis","Edad":19,"Altura":1.81},
                     {"Nombre":"Marta","Edad":33,"Altura":1.59},
                     {"Nombre":"Julian","Edad":48,"Altura":1.71},
                     {"Nombre":"Nerea","Edad":61,"Altura":1.78},
])
persona_min = gent.loc[gent["Altura"].idxmin()]
print(persona_min)
```

Promig edats i arrodonir a 2 decimals

```python
import pandas as pd
gent = pd.DataFrame([{"Nombre":"Pedro","Edad":25,"Altura":1.68},
                     {"Nombre":"Sonia","Edad":52,"Altura":1.74},
                     {"Nombre":"Lluis","Edad":19,"Altura":1.81},
                     {"Nombre":"Marta","Edad":33,"Altura":1.59},
                     {"Nombre":"Julian","Edad":48,"Altura":1.71},
                     {"Nombre":"Nerea","Edad":61,"Altura":1.78},
])
edad_promig = gent["Edad"].mean().round(2)
print(edad_promig)
```
